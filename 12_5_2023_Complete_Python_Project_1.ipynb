**Problem Statement:** Identify two Brooklyn based high schools and compare test performance results for the 2017 Global History & Geography regents exam

---



**Step 1**-**Data Cleaning**: Clean the data of the nonsensical value **'s'** and remove non-relevant columns.Obtain a general overview of the data source.



```python
missing_values=['s', np.nan]
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv', na_values=missing_values)
```

```python
df.head()
```

```python
df.tail()
```

```python
columns13 = df.drop(['Number Scoring CR','Percent Scoring CR'] , axis=1)
columns13
```

```python
dropval= columns13.dropna()
dropval
```

```python
dropval.columns
```

```python
display(dropval.dtypes)
```

```python
dfintegers = dropval.astype({"Number Scoring Below 65":'int64', "Number Scoring 65 or Above":'int64',"Number Scoring 80 or Above":'int64'})
```

```python
dfintegers
```

```python
display(dfintegers.dtypes)
```

**Note:** DBN each letter is for a specific borough
Brooklyn - K
Bronx- X
Queens - Q
Manhattan - M
Staten Island - R

```python
brooklyn_schools = dfintegers[df['School DBN'].str.contains('K')]
brooklyn_schools
```

```python
HS2017 = dfintegers[(brooklyn_schools['School Level'] == 'High school') & (df['Year'] == 2017)]
HS2017.head(3)
```

```python
HS2017.groupby(['School DBN', 'School Name','Regents Exam','Year'])['Total Tested'].sum()
```

```python
filt= dfintegers[(HS2017['School DBN'] == '13K350') | (df['School DBN'] == '84K739')]
```

```python
filt
```

```python
exam = ['Global History and Geography']
exam_type = filt[dfintegers['Regents Exam'].isin(exam)]
```

```python
exam_type.head()
```

```python
visual= exam_type.groupby(['School Name'])['Total Tested'].sum()
visual
```

```python
ax = visual.plot(x='School Name', y='Total Tested', kind='bar', legend=False, rot=0,
                 title='Students tested for 2017 Global History & Geography regents exam',
                 xlabel='School Name',
                 ylabel='Total Tested',
                 fontsize=8)
ax.bar_label(ax.containers[0], label_type='edge')
plt.tight_layout()
```

```python
agg1 = {'Number Scoring Below 65': 'sum','Number Scoring 65 or Above' : 'sum', 'Number Scoring 80 or Above' : 'sum', 'Total Tested' : 'sum'}
df_new = exam_type.groupby(dfintegers['School Name']).aggregate(agg1)
```

```python
df_new
```

```python
percentage = {'School Name':['New Visions Charter High School for the Humanities'],'Number Scoring Below 65' : [354],'Number Scoring 65 or Above' : [307],'Number Scoring 80 or Above' : [58],'Total Tested' : [661]}

df_vision = pd.DataFrame(percentage)
print(df_vision)
```

```python
df_vision[['percent below 65','percent 65 or above','percent 80 or above']] = (df_vision[['Number Scoring Below 65','Number Scoring 65 or Above','Number Scoring 80 or Above']] / df_vision['Total Tested'].values[:,None]) * 100
```

```python
df_vision
```

```python
round1= df_vision.round()
round1
```

```python
display(round1.dtypes)
```

```python
vision_int = round1.astype({"percent below 65":'int64', "percent 65 or above":'int64',"percent 80 or above":'int64'})
vision_int
```
