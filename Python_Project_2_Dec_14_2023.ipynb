```python
# Drop column 'weight'
df2= df.drop(['weight'] , axis=1)
df2
```

```python
# Total count of gender in the dataset
df2.groupby(['gender'])['gender'].count().sort_index(ascending=True)
```

```python
# Total count by race & gender in the dataset
df2.groupby(['race','gender'])['race'].count().sort_index(ascending=True)
```

'diabetesMed' column use to find how many people take diabetes medication 'YES' and were readmitted to hospital. Conclusion would be that those who take the meds have a lower readmittance rate. Can also compare with 'insulin;



```python
# Filter by specific value(s)
filter1 = df2[(df['diabetesMed'] == 'Yes')]
filter1.head(5)
```

```python
# Filter out records where race is indicated as '?'
rc = ['Asian','AfricanAmerican','Caucasian','Hispanic','Other']
filter2 = filter1[df2['race'].isin(rc)]
```

```python
# Total diabetesMed by specific race
subset1 = filter2.groupby(['race'])['diabetesMed'].count().sort_index(ascending=True)
subset1
```

```python
# Bar chart of total number of racial demographics who take diabetesMed
ax = subset1.sort_values(ascending=False).plot(kind='bar',color='pink')
ax.set(xlabel='Diabetes Medication', ylabel='Race', title='Total number of Diabetes Medication taken by race')
ax.bar_label(ax.containers[0], label_type='edge')
```

```python
# Total number of diabetesMed taken by age group in dataset
subset2 = filter1.groupby(['age'])['diabetesMed'].count()
subset2
```

```python
# Pie chart of total number of diabetesMed taken by age grounp
ax = subset2.sort_values(ascending=False).plot(kind='barh',color=['blue','red','yellow','orange','purple','magenta','grey','green','black','pink'])
ax.set(xlabel='Diabetes Medication', ylabel='Age Group', title='Total number of Diabetes Medication taken by age group')
ax.bar_label(ax.containers[0], label_type='edge')
```

```python
# Find number of medications prescribed to patients taking diabetes medication by age
subset3= filter1.groupby(['age'])['num_medications'].describe()
subset3
```
